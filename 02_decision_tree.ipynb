````python
# Packages
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import time
from sklearn.tree import DecisionTreeClassifier, plot_tree

# Pipeline
from tree import TreeModel

# ===== import data; tr,ts ======
X_tr = pd.read_csv('split-data/processed/X_tr.csv')
X_ts = pd.read_csv('split-data/processed/X_ts.csv')
y_tr = pd.read_csv('split-data/processed/y_tr.csv')
y_ts = pd.read_csv('split-data/processed/y_ts.csv')
````

# Balanced

No reweighting. Model learns with raw, possibly imbalanced data.

````python
# ===== Decision Tree ======
start = time.time()

dt_params = {
    'X_tr': X_tr.to_numpy(),
    'X_ts': X_ts.to_numpy(),
    'y_tr': y_tr.to_numpy(),
    'y_ts': y_ts.to_numpy(),
    'model': DecisionTreeClassifier(),
    'name': 'Decision Tree'
}

dt_pipeline = TreeModel(**dt_params)

dt_fit_params = {
    'param_grid': {
        'max_depth': [2, 3, 5, 10, 20, 50, 75, 100],
        'min_samples_leaf': [1, 5, 10, 20, 50],
        'min_samples_split': [2, 5, 10, 20],
    },
    'balance_classes': False,
    'k_folds': 5,
    'scoring': 'accuracy'
}

dt_pipeline.fit_predict(**dt_fit_params)

end = time.time()
elapsed = end - start
print(f"Elapsed time: {elapsed:.2f} seconds")
````

````python
dt_pipeline.tr_ts_scoring

plot_vhgr_params = {
    'param_list':dt_pipeline.hyper_grid,
    'figsize':(5, 4),
    'save_path':'media/02_decision_trees/validation_curves.png'
}
dt_pipeline.plot_validation_hyperparam_grid_row(**plot_vhgr_params)

dt_pipeline.viz_heatmap_cv(top_two_params_for_overfitting=['max_depth','min_samples_leaf'], dir_save_path='media/02_decision_trees')
````

``` ['max_depth','min_samples_split']``` 
= features with overfitting properties. 

````python
best_tree = dt_pipeline.grid.best_estimator_

plt.figure(figsize=(20, 10))
plot_tree(best_tree, feature_names=X_tr.columns, class_names=True, filled=True)
plt.title("Best Decision Tree")
plt.savefig('media/02_decision_trees/decision_tree_viz.png')
````

````python
k = 10
tree_importances = pd.Series(best_tree.feature_importances_, index=X_tr.columns)
top_features = tree_importances.sort_values(ascending=False).head(k)

sns.barplot(
    x=top_features.sort_values().values,
    y=top_features.sort_values().index,
    hue=top_features.sort_values().values,
    dodge=False,
    palette='viridis',
    legend=False
)

plt.title(f"Top {k} Feature Importances - Decision Tree")
plt.xlabel("Feature Importance (Gini-based)")
plt.ylabel('')
plt.tight_layout()
plt.savefig('media/02_decision_trees/significant_features_viz.png')
plt.show()
````

# Unbalanced

Use automatic reweighting to give more importance to rare classes.

````python
# ===== Decision Tree ======
start = time.time()

dt_params = {
    'X_tr': X_tr.to_numpy(),
    'X_ts': X_ts.to_numpy(),
    'y_tr': y_tr.to_numpy(),
    'y_ts': y_ts.to_numpy(),
    'model': DecisionTreeClassifier(),
    'name': 'Decision Tree'
}

dt_pipeline = TreeModel(**dt_params)

dt_fit_params = {
    'param_grid': {
        'max_depth': [2, 3, 5, 10, 20, 50, 75, 100],
        'min_samples_leaf': [1, 5, 10, 20, 50],
        'min_samples_split': [2, 5, 10, 20],
    },
    'balance_classes': True,
    'k_folds': 5,
    'scoring': 'accuracy'
}

dt_pipeline.fit_predict(**dt_fit_params)

end = time.time()
elapsed = end - start
print(f"Elapsed time: {elapsed:.2f} seconds")
````

````python
dt_pipeline.tr_ts_scoring

plot_vhgr_params = {
    'param_list':dt_pipeline.hyper_grid,
    'figsize':(5, 4),
    'save_path':'media/02_decision_trees/validation_curves.png'
}
dt_pipeline.plot_validation_hyperparam_grid_row(**plot_vhgr_params)

dt_pipeline.viz_heatmap_cv(top_two_params_for_overfitting=['max_depth','min_samples_leaf'], dir_save_path='media/02_decision_trees')
````

````python
best_tree = dt_pipeline.grid.best_estimator_

plt.figure(figsize=(20, 10))
plot_tree(best_tree, feature_names=X_tr.columns, class_names=True, filled=True)
plt.title("Best Decision Tree")
plt.savefig('media/02_decision_trees/decision_tree_viz.png')
````

````python
k = 10
tree_importances = pd.Series(best_tree.feature_importances_, index=X_tr.columns)
top_features = tree_importances.sort_values(ascending=False).head(k)

sns.barplot(
    x=top_features.sort_values().values,
    y=top_features.sort_values().index,
    hue=top_features.sort_values().values,
    dodge=False,
    palette='viridis',
    legend=False
)

plt.title(f"Top {k} Feature Importances - Decision Tree")
plt.xlabel("Feature Importance (Gini-based)")
plt.ylabel('')
plt.tight_layout()
plt.savefig('media/02_decision_trees/significant_features_viz.png')
plt.show()
````

To optimize classification performance using scoring='accuracy', we chose not to weigh the minority class higher. This prioritizes overall accuracy. However, if misclassification of the minority class were more critical, we would use class weighting and evaluate with a metric like recall instead.
